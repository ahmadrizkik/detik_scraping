```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
```

```python
text = open("berita.txt", "r").read().upper()
with open("stopwords_id.txt", "r") as stopwords:
    stopwords ={item.replace("\n", "") for item in stopwords.readlines()}
stopwords.update(["vs"])
mask = np.array(Image.open("mask.jpg"))
font = "montserrat/static/Montserrat-Bold.ttf"
```

```python
# masking
wordcloud = WordCloud(
                height=720, 
                width=1280, 
                stopwords=stopwords, 
                background_color="black",
                font_path=font,
                mask=mask,
                min_font_size=4)
```

```python
wordcloud.generate(text)
plt.figure(figsize=(9, 16))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(                
                height=720, 
                width=1280,
                stopwords=stopwords,
                font_path=font,
                background_color="white",
                min_font_size=16)
```

```python
wordcloud.generate(text)
plt.figure(figsize=(9, 16))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("wordcloud.png")
```
